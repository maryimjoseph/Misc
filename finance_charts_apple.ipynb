```python
# Please also add the actual day of the entry into its own column (e.g. Monday, Tuesday, etc.)

df['Day of Week'] = pd.to_datetime(df['Date']).dt.day_name()
df = df[['Date', 'Day of Week', 'AAPL.Open', 'AAPL.High', 'AAPL.Low', 'AAPL.Close',
       'AAPL.Volume', 'AAPL.Adjusted', 'dn', 'mavg', 'up', 'direction']]
df
```

3. Please aggregate the data sets to a week level and save the new data set as a
separate csv file

```python
# Step 1 : Add supporting columns to aggregate the data sets to a week level

df['Year'] = pd.to_datetime(df['Date']).dt.year
df['Week Number'] = pd.to_datetime(df['Date']).dt.isocalendar().week
df = df[['Date', 'Day of Week', 'Year', 'Week Number', 'AAPL.Open', 'AAPL.High', 'AAPL.Low',
       'AAPL.Close', 'AAPL.Volume', 'AAPL.Adjusted', 'dn', 'mavg', 'up',
       'direction']]
df.head(14)
```

```python
# Step 2 : aggregate the data sets to a week level

dfAggregates = df.groupby(['Year','Week Number'], axis=0).agg('mean',numeric_only=True)
dfAggregates
```

```python
# Step 3 : save the new data set as a separate csv file
dfAggregates.to_csv('finance-apple-weekly-aggregates.csv', header=True)
new_df = pd.read_csv('finance-apple-weekly-aggregates.csv')
new_df
```

**4. Please graph the results from step #3 visually as a candlestick chart (you may utilize any open-source library for this portion of the exercise)**

```python
import plotly.graph_objects as go

dfForGraph = pd.read_csv('/content/finance-apple-weekly-aggregates.csv')

dfForGraph['Year_Week'] = dfForGraph['Year'].apply(str) + '_' + dfForGraph['Week Number'].apply(str)
dfForGraph = dfForGraph[['Year_Week', 'Year', 'Week Number', 'AAPL.Open', 'AAPL.High', 'AAPL.Low','AAPL.Close',
                         'AAPL.Volume', 'AAPL.Adjusted', 'dn', 'mavg', 'up']]

graph = go.Figure(data=[go.Candlestick(x=dfForGraph['Year_Week'],
                          open=dfForGraph['AAPL.Open'], high=dfForGraph['AAPL.High'],
                          low=dfForGraph['AAPL.Low'], close=dfForGraph['AAPL.Close'],
                                       increasing_line_color='green', decreasing_line_color='orange')])

graph.update_layout(
  title='Weekly Performance of Apple Stock between February 17, 2015 and February 16, 2017',
  yaxis_title='AAPL Stock Price'
)
graph.show()
```
